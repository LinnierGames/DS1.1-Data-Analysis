```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x_1 = np.random.normal(-5, 1, 3000)
x_2 = np.random.normal(2, 3, 7000)
tupleeee = [x_1, x_2]
x = np.concatenate(tupleeee)
```

```python
import matplotlib.pyplot as plt

plt.hist(x, bins=20, density=2)
```

```python


plt.hist(x_1, bins=20, range=(-10, 10))
```

```python

plt.hist(x_2, bins=20, range=(-10, 10))
```

```python
import sklearn.mixture as mixture
gmm = mixture.GaussianMixture(n_components=2)
gmm.fit(x.reshape(-1,1))
print(gmm.means_)
print(gmm.covariances_)
print(gmm.weights_)
```

```python
z = gmm.sample(10_000)

plt.hist(z[0], bins=20, density=1)
```

# Kernel Density

```python
from sklearn.neighbors import KernelDensity

kde = KernelDensity(kernel='gaussian', bandwidth=0.6)
kde.fit(x.reshape(-1,1))

x_min, x_max = np.min(x), np.max(x)
s = np.linspace(x_min, x_max)
log_pdf = kde.score_samples(s.reshape(-1,1))
plt.plot(s, np.exp(log_pdf))
```

```python
m = kde.sample(10_000)
plt.hist(m, bins=20, density=1)
plt.show()
```

# Correlation

```python
df = pd.read_csv("datasets/Iris.csv")
df.drop(columns=['Id', "Species"], inplace=True)
display(df.head(1))

# Compute correlation among all features
df_corr = df.corr()
display(df_corr)

# Plot the correlation
display(sns.heatmap(df_corr))
#display(sns.corrplot(df))

# function that computes the correlation
def corr(df):
    """
    df: is a data frame to correlate
    
    returns: a new dataframe with correlation values
    """
    columns = np.array(df.columns)
    values = np.array(df.values)
    
    result = []
    
    for i in range(len(columns)):
        for j in range(len(columns)):
            i_label = columns[i]
            i_values = values[i]
            j_label = columns[i]
            j_values = values[i]
            
            i_j_r = corr_for(i_values, j_values)
            result.append(i_j_r)
            
    return result
            

def corr_for(x, y):
    """
    :x a np.array of values
    :y a np.array of values
    
    :returns look here https://en.wikipedia.org/wiki/Pearson_correlation_coefficient
    """
    assert(len(x) == len(y))
    size = len(x)
    
    x_mean = x.mean()
    y_mean = y.mean()
    numerator = 0
    denominator_a = 0
    denominator_b = 0
    
    for i in range(size):
        numerator += (x[i] - x_mean) * (y[i] - y_mean)
        
    for i in range(size):
        denominator_a += pow(x[i] - x_mean, 2)
    denominator_a = np.sqrt(denominator_a)
    
    for i in range(size):
        denominator_b += pow(y[i] - y_mean, 2)
    denominator_b = np.sqrt(denominator_b)
    
    return numerator / (denominator_a * denominator_b)

df = pd.DataFrame({'A': [1, 3, 5, 7, 9], 'B': [0, 2, 4, 6, 8]})
display(corr(df))
    
```
