```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
DATA_SET = 'titanic.csv'
df = pd.read_csv(DATA_SET)
```

```python
# How many passanagers embarked from each desitnation
sns.countplot(x=df.Embarked, data=df)
```

```python
# hwat proption of men vs women survived
sns.barplot(x=df.Sex, y=df.Survived, data=df)
```

```python
deck_list = list()
for a_cabin in df.Cabin:
    if type(a_cabin) is str:
        deck_list.append(a_cabin[0])
    else:
        deck_list.append(np.nan)
        
df['Deck'] = deck_list
```

```python
sns.countplot(
    x=df.Deck,
    data=df
)
```
