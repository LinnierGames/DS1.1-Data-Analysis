```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
```

```python
type(df)
```

```python
cabin_series = df.Cabin

type(cabin_series)
```

```python
deck_list = list()
for a_cabin in cabin_series:
    if type(a_cabin) is str:
        deck_list.append(a_cabin[0])
    else:
        deck_list.append(np.nan)
        
deck_list
```

```python
df['Deck'] = deck_list

df.head()
```

```python
df[(df.Deck == "A") | (df.Deck == "B") & (df.Survived == 1)].describe()
```

```python
df = df.dropna()
```

```python
count_df = len(df)
count_df
```

```python
survivors_df = df[df.Survived == 1]
len(survivors_df)
```

```python
youngest_age = survivors_df.Age.min()
youngest_age
```

```python
youngest_df = survivors_df.Age.sort_values()
youngest_df
```
