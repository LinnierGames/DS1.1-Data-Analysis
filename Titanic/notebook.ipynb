```python
# Slice the data frame and include only records Embarked is "C" and Sex is "male"
titanic_df[(titanic_df.Embarked == "C") & (titanic_df.Sex == "male")].head()
```

```python
titanic_df.describe()
```

```python
men_df = titanic_df[(titanic_df.Sex == "male")]
men_who_died_df = titanic_df[(titanic_df.Survived == 0)]
men_who_lived_df = titanic_df[(titanic_df.Survived == 1)]

survied_df = titanic_df[(titanic_df.Survived == 1)]
print("Survied: {}".format(len(survied_df.index)))
```

```python
mean_ticket = survied_df.Fare.mean()
print(mean_ticket)
```

```python
who_survived_top_fare_df = survied_df[survied_df.Fare >= mean_ticket]
len(who_survived_top_fare_df.index)
print("Survied top half of fare: {}".format(len(who_survived_top_fare_df.index)))
```

```python
who_survived_bottom_fare_df = survied_df[survied_df.Fare < mean_ticket]
print("Survied top half of fare: {}".format(len(who_survived_bottom_fare_df.index)))
```

```python
sets = titanic_df[(titanic_df.Sex == "male") & (titanic_df.Embarked == "C")]
a = sets.Sex == "male"

sns.barplot(a)
```
