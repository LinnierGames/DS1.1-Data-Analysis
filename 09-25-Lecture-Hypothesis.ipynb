```python
import pandas as pd
import numpy as np
```

```python
from scipy.stats import norm

mean = 5
std_div = 2
df = pd.DataFrame({"point": norm.rvs(loc=mean, scale=std_div, size=100)})
```

```python
df.plot()
```

```python
def z_index(value):
    return(value - mean) / std_div
    
df = df.apply(z_index)
```

```python

# Rat Example
def reject_null_hpy(pop_mean, sample_mean, size, pop_std_deviation, sig_level):
    z = ( sample_mean - pop_mean ) / (pop_std_deviation / np.sqrt(size))
    p = 2 * norm.cdf(-np.abs(z))
    
   #  display(z, p)
    
    display(2 * norm.cdf(-np.abs(z)), 2 * norm.sf(np.abs(z)))
    
    if p < sig_level:
        return True
    else:
        return False
    
reject_null_hpy(1.2, 1.05, 100, 0.5, 0.05)
```

```python
x = [177.3, 182.7, 169.6, 176.3, 180.3, 179.4, 178.5, 177.2, 181.8, 176.5]
x_size = len(x)
x_mean = np.mean(x)
x_std = np.std(x, ddof=1)

def reject_null_hpy(data, pop_mean, sample_mean, size, sample_std_deviation):
    t_score = ( sample_mean - pop_mean ) / (sample_std_deviation / np.sqrt(size))
    t = stats.ttest_1samp(data, pop_mean)

    display(t, t_score)

reject_null_hpy(x, 175.3, x_mean, x_size, x_std)
```
