```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x_1 = np.random.normal(-5, 1, 3000)
x_2 = np.random.normal(2, 3, 7000)
tupleeee = [x_1, x_2]
x = np.concatenate(tupleeee)
```

```python
import matplotlib.pyplot as plt

plt.hist(x, bins=20, density=2)
```

```python


plt.hist(x_1, bins=20, range=(-10, 10))
```

```python

plt.hist(x_2, bins=20, range=(-10, 10))
```

```python
import sklearn.mixture as mixture
gmm = mixture.GaussianMixture(n_components=2)
gmm.fit(x.reshape(-1,1))
print(gmm.means_)
print(gmm.covariances_)
print(gmm.weights_)
```

```python
z = gmm.sample(10_000)

plt.hist(z[0], bins=20, density=1)
```

# Kernel Density

```python
from sklearn.neighbors import KernelDensity

kde = KernelDensity(kernel='gaussian', bandwidth=0.6)
kde.fit(x.reshape(-1,1))

x_min, x_max = np.min(x), np.max(x)
s = np.linspace(x_min, x_max)
log_pdf = kde.score_samples(s.reshape(-1,1))
plt.plot(s, np.exp(log_pdf))
```

```python
m = kde.sample(10_000)
plt.hist(m, bins=20, density=1)
plt.show()
```
